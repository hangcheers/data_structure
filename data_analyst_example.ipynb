```python
pd.pivot_table(df, index = [u'主客场',u'胜负'], values = [u'得分',u'助攻',u'篮板'], aggfunc = [np.sum, np.mean] )
```

```python
df1 = pd.DataFrame({'name':['w','z','t','l'],'Chinese':[95,88,23,66],
                   'Politics':[96,64,33,66],'English':[66,100,33,66]},index = [0,1,2,4])
print(df1)
```

```python
df2 = pd.DataFrame({'name':['h','k','d','a'],'English':[98,95,94,67],
                   'Basketball':[7,10,13,8],'Failure':[10,2,0,8]},index = [1,2,3,5])
print(df2)
```

```python
frame = [df1,df2]
df3 = pd.concat(frame,keys=['1','2'],axis = 0, join = 'outer', sort = True).fillna(0)
print(df3)
```

```python
df4 = pd.concat(frame, keys = ['1','2'], axis = 1, join = 'inner', sort = True)
print(df4)
```
